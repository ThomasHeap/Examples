## Randomised prior functions

Uncertainty should come both from the data and from the prior. By adding a randomly sampled prior to each member in the ensemble we should be able to account for this.

Implementation in pytorch following: https://gdmarmerola.github.io/intro-randomized-prior-functions/
Borrowing his plotting functions liberally

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import tqdm
```

### Data Generating function

```python
from torch.distributions import Uniform, Normal


def data_fn(num_samples):
    unif = Uniform(0.0, 0.5)
    normal = Normal(0,0.02)
    
    x = unif.sample((num_samples,))
    eps = normal.sample((num_samples,))
    y = x + 0.3*torch.sin(2*np.pi*(x+eps)) + 0.3*torch.sin(4*np.pi*(x+eps)) + eps - 0.5
    return x,y
```

```python
x,y = data_fn(100)

fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(x, y, c='black', marker='x')

plt.xlim([-1.5,2])
plt.ylim([-2,2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Data Generating process')
fig.tight_layout()
```

### Prior network

```python
from torch import nn

x_grid = torch.linspace(-5, 5, 1000).reshape(-1,1)

prior_net = nn.Sequential(
            nn.Linear(1,16),
            nn.ELU(),
            nn.Linear(16,16),
            nn.ELU(),
            nn.Linear(16,1))

class Prior(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1 = nn.Linear(1,16)
        self.l2 = nn.Linear(16,16)
        self.l3 = nn.Linear(16,1)
        self.elu = nn.ELU()
        
        
    def forward(self, x):
        x = self.l1(x)
        x = self.elu(x)
        x = self.l2(x)
        x = self.elu(x)
        
        return self.l3(x)

def init_weights(m):
    if isinstance(m, nn.Linear):
        torch.nn.init.xavier_normal_(m.weight)
        m.bias.data.fill_(0)
        
```

```python
prior_net = Prior()
prior_net.apply(init_weights)

prior_preds = prior_net(x_grid)

x,y = data_fn(100)

fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(x, y, c='black', marker='x')
ax.plot(x_grid.detach().numpy(),prior_preds.detach().numpy())
plt.xlim([-0.5,1])
plt.ylim([-1,1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Data Generating process + prior')
fig.tight_layout()
```

### Trainable network

```python
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1 = nn.Linear(1,16)
        self.l2 = nn.Linear(16,16)
        self.l3 = nn.Linear(16,1)
        self.elu = nn.ELU()
        
        
    def forward(self, x):
        x = self.l1(x)
        x = self.elu(x)
        x = self.l2(x)
        x = self.elu(x)
        
        return self.l3(x)
```

### Fitting Model

```python
trainable_net = Net()

mse = nn.MSELoss()
optimizer = torch.optim.Adam(trainable_net.parameters(), lr=0.001)

epochs = tqdm.tqdm(range(2000))

for i in epochs:
    
    prior_out = prior_net(x.reshape(-1,1))
    trainable_out = trainable_net(x.reshape(-1,1))
    out = trainable_out + prior_out
    
    optimizer.zero_grad()
    
    loss = mse(out,y.reshape(-1,1))
    
    loss.backward()
    optimizer.step()
    epochs.set_description(f"Loss: {loss}")
    
trainable_preds = trainable_net(x_grid)
resultant = trainable_preds + prior_preds
```

```python
x,y = data_fn(100)

fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(x, y, c='black', marker='x')
ax.plot(x_grid.detach().numpy(),prior_preds.detach().numpy(), label = 'Prior')
ax.plot(x_grid.detach().numpy(),trainable_preds.detach().numpy(), label = 'Net')
ax.plot(x_grid.detach().numpy(),resultant.detach().numpy(), label = 'Prior + Net')

plt.xlim([-0.5,1])
plt.ylim([-1,1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Data Generating process + prior + trainable net')
ax.legend()
fig.tight_layout()
```

### Randomized Priors

There is probably a better parallelised way of doing this.

```python
prior_preds = []
trainable_preds = []
resultants = []

for i in range(9):
    prior_net = Prior()
    prior_net.apply(init_weights)

    p_preds = prior_net(x_grid)
    trainable_net = Net()

    mse = nn.MSELoss()
    optimizer = torch.optim.Adam(trainable_net.parameters(), lr=0.001)

    epochs = tqdm.tqdm(range(3000))

    for i in epochs:

        prior_out = prior_net(x.reshape(-1,1))
        trainable_out = trainable_net(x.reshape(-1,1))
        out = trainable_out + prior_out

        optimizer.zero_grad()

        loss = mse(out,y.reshape(-1,1))

        loss.backward()
        optimizer.step()
        epochs.set_description(f"Loss: {loss}")

    t_preds = trainable_net(x_grid)
    resultant = t_preds + p_preds
    
    prior_preds.append(p_preds.detach())
    trainable_preds.append(t_preds.detach())
    resultants.append(resultant.detach()) 
```

```python
#plotting

plt.figure(figsize=[16,12], dpi=200)

# loop for plotting predictions of each head 
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.plot(x, y, 'kx', label='Toy data', alpha=0.8)
    plt.plot(x_grid.detach().numpy(), prior_preds[i].numpy(), label='prior net (p)')
    plt.plot(x_grid.detach().numpy(), trainable_preds[i].numpy(), label='trainable net (f)')
    plt.plot(x_grid.detach().numpy(), resultants[i].numpy(), label='resultant (Q)')
    plt.title('Ensemble: Model #{}'.format(i+1), fontsize=14)
    plt.xlabel('$x$'); plt.ylabel('$y$')
    plt.xlim(-0.5,1.0); plt.ylim(-1.0,1.0)
plt.tight_layout()
plt.show();
```

```python
# computing mean and stddev
mean = np.array(torch.stack(resultants)).mean(axis=0)
std = np.array(torch.stack(resultants)).std(axis=0)

# opening figure
fig = plt.figure(figsize=[12,5], dpi=150)

# title of the plot
fig.suptitle('Uncertainty estimates given by bootstrapped ensemble of neural networks with randomized priors', verticalalignment='center')

# first subplot, 
plt.subplot(1, 2, 1)

# let us plot the training data
plt.plot(x, y, 'kx',  label='Toy data')
plt.title('Mean and Deviation', fontsize=12)
plt.xlim(-0.5, 1.0); plt.ylim(-1.5, 1.5)
plt.legend()

# plotting predictive mean and deviation
plt.plot(x_grid, mean, 'r--', linewidth=1.5)
plt.fill_between(x_grid.reshape(1,-1)[0], (mean - std).reshape(1,-1)[0], (mean + std).reshape(1,-1)[0], alpha=0.5, color='red')
plt.fill_between(x_grid.reshape(1,-1)[0], (mean + 2*std).reshape(1,-1)[0], (mean - 2*std).reshape(1,-1)[0], alpha=0.2, color='red')

# second subplot
plt.subplot(1, 2, 2)

# let us plot the training data
plt.plot(x, y, 'kx', label='Toy data')
plt.title('Samples', fontsize=12)
plt.xlim(-0.5, 1.0); plt.ylim(-1.5, 1.5)
plt.legend()

# loop for plotting predictions of each head 
for i in range(9):
    plt.plot(x_grid, resultants[i].numpy(), linestyle='--', linewidth=1.5)
    
# showing
plt.show();
```
